```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
import time

n = 5000
m = 300
kList = []
k = 0
domainList = [0]*n
complete = False
while not complete:
    k += 1
    randNum = random.randint(0,4999)

    if domainList[randNum] == 1:
        complete = True
    else:
        domainList[randNum] = 1

print(k)

startTime = time.time()
for i in range(m):
    domainList = [0]*n
    k = 0
    complete = False
    while not complete:
        k += 1
        randNum = random.randint(0,4999)

        if domainList[randNum] == 1:
            complete = True
        else:
            domainList[randNum] = 1
    kList.append(k)
finishTime = time.time() - startTime
sns.kdeplot(kList)
plt.yscale("linear")
plt.xscale('linear')
print(sum(kList)/len(kList))
print("300 trials took ", finishTime, "seconds")
```

#### How many trials did it take? (what is k)
It took 40 random trials until it found a match. It varies everytime you run the code. After repeating the experiment 300 times I found that the average of k was usually around 87. This means that, on average it takes 87 attemps to find a matching number in the domain.

#### How I Implemented the Experiment
I implemented this experiment in a way similar to how we did lab 1. I created a list that I made the size of the domain, n, which was 5000. Then I randomly generated numbers in that domain until a number in that domain had been generated twice. The numbers spot in the list would be marked one if it was already generated and 0 if it had not been. I had to make the random number in a range of 0 to n - 1 since lists start at 0 and not at 1.

## Runtime Test

```python
n = 5000
m = 300
runTimeList = []

while m < 50300:
    startTime = time.time()
    for i in range(m):
        domainList = [0]*n
        complete = False
        while not complete:
            randNum = random.randint(0,n-1)

            if domainList[randNum] == 1:
                complete = True
            else:
                domainList[randNum] = 1
    finishTime = time.time() - startTime
    print("At n = ", n, "m = ", m, " the runtime was ", finishTime)
    runTimeList.append(finishTime)
    n += 2000
    m += 2000

plt.plot(runTimeList)
plt.title("Increasing N and M by 2000 Each Iteration")
plt.ylabel("Execution Time")
plt.xlabel("Iterations of increasing N and M")
```
