```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
import time

n = 300
m = 400
kList = []

complete = False
k = 0
domainList = [0]*n
while not complete:
    randNum = random.randint(0,299)
    if domainList[randNum] == 0:
        domainList[randNum] = 1

    k += 1

    if domainList.count(1) == len(domainList):
        complete = True

print(k)

startTime = time.time()
for i in range(m):
    complete = False
    k = 0
    domainList = [0]*n
    while not complete:
        randNum = random.randint(0,299)
        if domainList[randNum] == 0:
            domainList[randNum] = 1

        k += 1

        if domainList.count(1) == len(domainList):
            complete = True
    
    kList.append(k)
finishTime = time.time() - startTime
print(finishTime)
repetitionNum = [x + 1 for x in range(m)]
x = repetitionNum
y = kList

sns.kdeplot(kList)
plt.xlabel("Attemps")
```

#### How I Implemented This Experiment
To implement this experiment I did something similar to lab 1 but instead of looking to see if the number generated has already been generated, I have it so it checks if its spot on the list is equal to a 1, meaning it has already been generated, or a 0, meaning it hasn't been generated, and if it's a 0 I set it to 1. After that is checked I also check if every index in the list is equal to one using the count function. If every index is equal to one then the experiment is complete and the loop ends. During every iteration of the loop I iterate k by 1 to keep track of how many tries it takes before the list is filled.

## Runtime Test

```python
startTime = time.time()
runTimeList = []
n = 300
m = 400

while (n < 2000):
    startTime = time.time()
    for i in range(m):
        complete = False
        domainList = [0]*n
        while not complete:
            randNum = random.randint(0, n-1)
            if domainList[randNum] == 0:
                domainList[randNum] = 1

            if domainList.count(1) == len(domainList):
                complete = True
    finishTime = time.time() - startTime
    print("At n = ", n, " and m = ", m, " the runtime was ", finishTime)
    runTimeList.append(finishTime)
    n += 50
    m += 50
print(runTimeList)
plt.plot(runTimeList)
plt.title("Increasing N and M by 50 Each Iteration")
plt.ylabel("Execution Time")
plt.xlabel("Iterations of increasing N and M")
```

#### Runtime Graph
For my graph I increase n and m by 50 every iteration starting at n = 300 and m = 400. I had the limit of n < 2000. This quickly became a long wait for each iteration.
