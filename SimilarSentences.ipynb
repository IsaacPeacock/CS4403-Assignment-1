```python
import matplotlib.pyplot as plt
import seaborn as sns

# from your hashing notes
def shingle(text: str, k: int):
    shingle_set = []

    for i in range(len(text) - k+1):
        shingle_set.append(text[i:i+k])
    return set(shingle_set)

#help from ChatGPT for this function
def shingleWords(text: str):
    shingle_set = []
    words = text.split()

    for i in range(len(words) - 1):
        shingle_set.append((words[i], words[i+1]))
    return set(shingle_set)

# from your hashing notes
def jaccard(a: set, b: set):
    return len(a.intersection(b)) / len(a.union(b))
```

# Character Based 3-grams

```python
with open('Random spanish sentences.txt') as f:
    sentences = [sentences.strip() for sentences in f.readlines()]

#to get rid of repeat sentences
uniqueSentences = set(sentences)
sentences = list(uniqueSentences)

k = 3

#stores all created k-grams in a list
kGrams = []
#stores lists of k-grams created from each sentence
threeGram= []
for x in sentences:
     threeGram.append(shingle(x.lower(), k))
     kGrams += shingle(x.lower(), k)
     
#removes duplicate k-grams
uniqueThreeKGrams = set(kGrams)
print(len(uniqueThreeKGrams))

similarityList = []
#loops through all sentence k-grams and runs the jaccard similarity on all pairs
for i in range(len(threeGram)):
    sentence1 = threeGram[i]
    for j in range(i+1, len(threeGram)):
        similarityList.append(jaccard(sentence1, threeGram[j]))

resultList = []       
for i in range(len(threeGram)):
    for j in range(i+1, len(threeGram)):
        score = jaccard(threeGram[i], threeGram[j])
        resultList.append((sentences[i], sentences[j], score))

for i in range(len(resultList)):
    for j in range(i+1, len(resultList)):
        if resultList[i][2] < resultList[j][2]:
            temp = resultList[i]
            resultList[i] = resultList[j]
            resultList[j] = temp

print(resultList[:5])
```
